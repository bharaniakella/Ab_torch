```python
import numpy
import torch
import torch.nn as nn
from matplotlib import pyplot as plt
```

```python
N=30
x=torch.randn(N,1)
y=x+torch.randn(N,1)/2
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
nn_reg=nn.Sequential(
    nn.Linear(1,1),
    nn.ReLU(),
    nn.Linear(1,1)
)
```

```python
lr=0.05
loss_fn=nn.MSELoss()
optimizer=torch.optim.SGD(nn_reg.parameters(),lr=lr)
```

```python
numepochs=500
losses=torch.zeros(numepochs)
```

```python
for epochi in range(numepochs):
  yhat=nn_reg(x)

  loss=loss_fn(yhat,y)
  losses[epochi]=loss

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
final_pred=nn_reg(x)
```

```python
final_loss=(final_pred-y).pow(2).mean()
final_loss
```

```python
plt.plot(losses.detach(),'o')
plt.plot(numepochs,final_loss.detach(),'o')
plt.show()
```
